```python
import optuna
from transformers import TrainingArguments, Trainer
from transformers import T5Tokenizer, T5ForConditionalGeneration
from datasets import Dataset

def objective(trial):

    # Define hyperparameters to optimize
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 5e-4)
    batch_size = trial.suggest_categorical('batch_size', [4, 8, 16])
    num_train_epochs = trial.suggest_int('num_train_epochs', 1, 10)
    weight_decay = trial.suggest_loguniform('weight_decay', 1e-6, 0.1)

    # Load model and tokenizer
    tokenizer = T5Tokenizer.from_pretrained('t5-small')
    model = T5ForConditionalGeneration.from_pretrained('t5-small')

    # Preprocessing function
    def preprocess(datas):
        inp = [x for x in datas['input']]
        out = [x for x in datas['output']]

        model_inp = tokenizer(inp, max_length=128, truncation=True, padding='max_length')

        with tokenizer.as_target_tokenizer():
            labels = tokenizer(out, max_length=128, truncation=True, padding='max_length')

        model_inp["labels"] = labels["input_ids"]
        return model_inp

    # Apply preprocessing to the dataset
    token_train_data = train_data.map(preprocess, batched=True)
    token_val_data = validation_data.map(preprocess, batched=True)

    # Define training arguments
    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=num_train_epochs,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        warmup_steps=500,
        weight_decay=weight_decay,
        logging_dir='./logs',
        logging_steps=10,
        evaluation_strategy="epoch",
        learning_rate=learning_rate,
        disable_tqdm=True,  # Disable tqdm progress bar for cleaner output
    )

    # Define the Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=token_train_data,
        eval_dataset=token_val_data,
    )

    # Train the model
    trainer.train()

    # Evaluate the model
    eval_results = trainer.evaluate(eval_dataset=token_val_data)

    # Use the evaluation loss as the objective value to minimize
    return eval_results['eval_loss']

# Create a study object to optimize the objective function
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=10)  # You can increase n_trials for better optimization

# Display the best hyperparameters
print("Best hyperparameters:", study.best_params)

# Train the final model using the best hyperparameters
best_params = study.best_params

final_training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=best_params['num_train_epochs'],
    per_device_train_batch_size=best_params['batch_size'],
    per_device_eval_batch_size=best_params['batch_size'],
    warmup_steps=500,
    weight_decay=best_params['weight_decay'],
    logging_dir='./logs',
    logging_steps=10,
    evaluation_strategy="epoch",
    learning_rate=best_params['learning_rate']
)

final_trainer = Trainer(
    model=model,
    args=final_training_args,
    train_dataset=token_train_data,
    eval_dataset=token_val_data
)

# Train and save the final model
final_trainer.train()
model.save_pretrained("./t5_nmap_model_optimized")
tokenizer.save_pretrained("./t5_nmap_tokenizer_optimized")
```
